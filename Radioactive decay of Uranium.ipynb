### Radioactive decay of Uranium

Consider 1 gram of uranium at the beginning of formation of the Earth. Obtain a plot of radioactive decay of 238U and 235U as a function of time,assuming their natural abundance of 99.275 % and 0.725% respectively. The half-lives of 238U and 235U are  6.0X 10^9and 9.4X10^8 years, respectively, and their molar masses are 238.05079 and 235.04393 gram, respectively. Take the current age of the Earth as4.5X10^9  years. Display the results as a 4-plot panel figure, with two as regular and two as semi-log plots.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Radioactivity():
    def __init__(self,m,M,NA):
        self.m = m
        self.M = M
        self.NA = NA
        
    def nuclei_number(self):
        Nn = self.m*self.NA/self.M    #Used Nn.m = NA.M
        return Nn 
    
    def undecayed_nuclei(self,half_life,t):
        No = self.nuclei_number()
        self.half_life = half_life
        self.t = t
        N = No*np.exp(-0.693*self.t/self.half_life)
        return N
    
        
```

```python
NA = 6.0221408*10**23 # Avogadro's number

m1 = 0.99275 # mass of U238 in gram in 1 gram of uranium
m2 = 1 - m1 # mass of U235 in gram in 1 gram of uranium

M1 = 238.05079 # molar mass of U238
M2 = 235.04393 # molar mass of U235

half_life1 = 6.0*10**9 # half life of U238
half_life2 = 9.4*10**8 # half life of U235

T = []
N1 = []
N2 = []

t = np.linspace(0, 4.5*10**9, 2000)

for i in t:
    T.append(i)
    U_238 = Radioactivity (m1, M1, NA)
    U_235 = Radioactivity (m2, M2, NA)
    
    N_U238 = U_238.undecayed_nuclei(half_life1, i)
    N1.append(N_U238)
    
    N_U235 = U_235.undecayed_nuclei(half_life2, i)
    N2.append(N_U235)
    
```

```python
plt.figure(figsize = [15,15])
plt.suptitle('Radioactive decay of Uranium',fontsize = 16)

plt.subplot(3,3,1)
plt.plot(T,N1,"g.",linewidth = 3)
plt.xlabel('Time(years)')
plt.ylabel('N')
plt.title('Radioactive decay of U-238', fontsize = 12)

plt.subplot(3,3,2)
plt.plot(T,N2,"r.",linewidth = 3)
plt.xlabel('Time(years)')
plt.ylabel('N2')
plt.title('Radioactive decay of U-235', fontsize = 12)

plt.subplot(3,3,3)
plt.semilogy(T,N1,"b_",linewidth = 3)
plt.xlabel('Time(years)')
plt.ylabel('N1')
plt.title('Radioactive decay of U-238', fontsize = 12)

plt.subplot(3,3,4)
plt.semilogy(T,N2,"m_",linewidth = 3)
plt.xlabel('Time(years)')
plt.ylabel('N2')
plt.title('Radioactive decay of U-235', fontsize = 12)

plt.subplots_adjust(wspace = 0.4, hspace = 0.4)
```
